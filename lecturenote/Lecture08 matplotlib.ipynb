## 축 범위

* `plt.xlim(xmin, xmax)`, `plt.ylim(ymin, ymax)`
* `plt.axis([xmin, xmax, ymin, ymax])`
* `plt.axis(option="auto")` : 옵션 지정
    * `"auto"` : 자동 설정
    * `"normal"` : 자동 설정
    * `"on"`: 축과 라벨 출력
    * `"off"`: 축과 라벨 미출력
    * `"equal"` : 각 축의 범위와 축의 스케일 동일하게 설정
    * `"scaled"` : 그래프의 차원과 동일하게 축의 스케일 설정
    * `"tight"` : 전체 데이터가 보이는 범위에 맞게 축 확장
    * `"image"` : 데이터 범위에 대해 축의 범위를 사용한 `scaled`
    * `"square"` : x축과 y축의 크기가 같도록 설정
    

image.png

```python
plt.plot(x, y)
plt.xlim(1, 31)
plt.ylim(0, 30);
```

```python
plt.plot(x, y)
plt.axis("square");
```

```python
plt.plot(x, y)
plt.axis("scaled");
# x축과 y축의 비율을 실제 비율과 같은 비율로 나오게 하고, 여백을 축소함.
```

```python
# 축의 범위 얻기
plt.plot(x, y)
plt.axis("scaled");

xmin, xmax = plt.xlim()
ymin, ymax = plt.ylim()
print(xmin, xmax, ymin, ymax)

xmin, xmax, ymin, ymax = plt.axis()
print(xmin, xmax, ymin, ymax)

x_range, y_range = plt.xlim(), plt.ylim()
print(x_range, y_range)
```

#  그래프 종류

| 함수 | 설명 |
|------|------|
| `plot` | 플롯 라인 또는 마커 |
| `bar` | 세로 막대 그래프 |
| `barh` | 가로 막대 그래프 |
| `boxplot` | 박스 플롯 |

## 막대그림

```python
?plt.bar
```

```python
x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(x, values)
plt.xticks(x, years)

plt.show()
```

```python
## 너비조정
# plt.bar(x, values, width=0.4)
# plt.bar(x, values, width=0.6)
# plt.bar(x, values, width=0.8)
plt.bar(x, values, width=1)
plt.xticks(x, years)

plt.show()
```

```python
# 막대 색상 변경 / ylim(0, 2000)으로 설정
plt.bar(x, values, width=0.4, color="r")
plt.xlabel("Year")
plt.ylabel("value")
plt.ylim(0, 2000);
```

```python
# 가로막대그래프
y = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.barh(y, values)
plt.yticks(y, years)

plt.show();
```

```python
# 무늬 주기
plt.bar(x, values, hatch=".") # 점 모양
```

```python
# 무늬 주기
width = 0.3
x = ['A', 'B', 'C']
y = [100, 60, 85]

fig, axes = plt.subplots(2, 2)
fig.set_size_inches((10,10))
fig.set_facecolor('white')
axes[0][0].bar(x, y, width=width, color='none', edgecolor='k', hatch='o') ## 조그만 원모양
axes[0][1].bar(x, y, width=width, color='none', edgecolor='r', hatch='O') ## 큰 원모양
axes[1][0].bar(x, y, width=width, color='none', edgecolor='b', hatch='oo') ## 패턴을 반복하면 좀 더 조밀하게 적용
axes[1][1].bar(x, y, width=width, color='none', edgecolor='green', hatch='//..') ## 다른 패턴과 혼용 가능
plt.show()
```

## 파이차트


```python
?plt.pie
```

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
# 라벨 붙이기 + 소수점 첫번째 자리까지 붙이고 % 표시하기
plt.pie(ratio, labels=labels, autopct='%.1f%%'); # %.2f%%
# autopct 부채꼴 안에 표시될 숫자의 형식, 소수점 한자리까지 표시
```

```python
# 시작위치 변경하고, 시계방향으로 그리기
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False);
```

```python
# 조각을 떼서 강조하기 'Banana'와 'Grapes' 영역에 대하여 반지름의 10%만큼 벗어나도록 설정
explode = [0, 0.30, 0, 0.10]

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode);
```

```python
# 레전드 따로 붙이기 loc로 위치 조정 (0.5,0.5)가 figure의 중앙
plt.pie(ratio, labels=labels)
plt.legend(loc=(1.0, 0.5), title="Fruits");
```
