## duplicated()

* `DataFrame.duplicated(subset=None, keep="first")`
    * 메소드가 적용된 열과 길이가 동일한 boolean 벡터를 반환
    * 어떤 요소가 중복인지 확인 가능
    * `subset` : 중복데이터를 확인할 열 지정. Series, index에 적용할 경우 사용하지 않음
    * `keep` : 중복데이터 발견 시 어떤 데이터를 마크할지 삭제할지 결정
        * `keep="first"(default)` : 처음에 발견된 데이터를 제외하고 중복인 경우 True
        * `keep="last"` : 마지막으로 발견된 데이터를 제외하고 중복인 경우 True
        * `keep=False` : 모든 중복데이터에 대해 True

* `DataFrame.drop_duplicates(subset=None, keep="first", inplace=False)`
    * `inplace=False` : 기존 데이터프레임 반영여부


````python
df = pd.DataFrame(data = [[1,2,3], [1,2,4], [1,3,3], [1,2,4]], columns=["a", "b", "c"])
df
````

````python
df.duplicated()  # 각 행별로 모든 요소가 중복인지 체크
#default(keep="first") 처음 발견된 데이터 제외
````

````python
print(df.duplicated("a", keep = 'first'))  # a열 중복체크
print(df.duplicated('a', keep = 'last'))
print(df.duplicated('a', keep = False))
````

````python
df.duplicated(["a", "b"])  # a,b열 중복체크
````

````python
df.duplicated(["a", "b"], keep="last")
````

````python
df.drop_duplicates()
````

````python
print(df)
df.drop_duplicates("a", inplace=True)
print(df)
````

## sort()

* `DataFrame.sort_values(by=None, axis=0, ascending=True, inplace=False, na_position="last")`
    * 해당 변수의 값 기준으로 정렬
    * `by=None` : 정렬할 기준 변수 지정
    * `axis` : `0` 또는 `"index"` 입력 시 행 기준 정렬. `1` 또는 `"columns"` 입력 시 열 기준 정렬
    * `ascending` : `True`이면 오름차순, `False`이면 내림차순
    * `na_position` : 결측값 위치 ("first", "last")

* `DataFrame.sort_index(axis=0, asceding=True, inplace=False, na_position="last"))`
    * 인덱스 기준으로 정렬

## rank()

* `rank(method="average")`
    * 값들의 순위 지정
    * `method` : 같은 값에 대한 처리방법
        * `method="average"` : 같은 값을 가지는 항목들의 순위 평균값 사용
        * `method="min"` : 같은 값을 가지는 항목들의 순위 최소값 사용
        * `method="max"` : 같은 값을 가지는 항목들의 순위 최대값 사용
        * `method="dense"` : 같은 값을 가지는 항목들의 순위를 동일하게 하고, 전체 순위를 1씩 증가
        * `method="first"` : 같은 값을 가지는 항목들의 순위를 데이터 내 위치에 따라 1씩 증가하게 부여


````python
a = pd.Series([1, 7, -3, 1, 1, 1])
pd.concat([a, a.rank()], axis = 1)
a
````

````python
a.rank(method="min")
````

````python
a.rank(method="max")
````

````python
a.rank(method="dense")
````

````python
a.rank(method="first")
````

````python
a.cummin() #누적최소값 <=> cummax() 누적최대값
# 누적된 값 중 최소값 or 최대값 반환
````

````python
a.cummax()
````

함수 | 설명
:--- | :---

## merge()

* `pd.merge(data1, data2,left_index=False, right_index=False, how="outer", left_on=None, right_on=None, suffixes=("_x", "_y")`
    * `left_index`, `right_index` : 데이터를 인덱스를 기준으로 행이나 열방향으로 붙임
    * `left_on`, `right_on` : 데이터를 특정 변수를 기준으로 병합. 여러 개 변수 지정 가능.
    * `how` : 병합 기준. "outer", "inner", "left", "right"


* `DataFrame.join(data2, on=None, how="left", lsuffix="", rsuffix="")`
    * 데이터프레임의 메소드
    * lsuffix, rsuffix: 접두어 붙이기(구분용)
